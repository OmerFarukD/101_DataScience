```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.style.use('seaborn-darkgrid')
```

```python
plt.figure(figsize=(8,6))
area=[48,30,25,20]
labels=["Low","Medium","High","Critical"]
colors=["#FFFFF0","#FF7F50","#006400","#00FFFF"]
plt.pie(area,labels=labels,colors=colors,startangle=45,shadow="true",autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(8,8))
area=[20,25,10,35]
labels=["Low","Medium","High","Critical"]
colors=['#7CB342', '#C0CA33', '#FFB300', '#F57C00']
total=np.sum(area)
def val_per(x):
    return '{:.2f}%\n({:.1f})'.format(x,total*x/100)
plt.pie(area,labels=labels,colors=colors, autopct=val_per)
plt.show()
```

```python
plt.figure(figsize=(8,8))
area=[10,16,12,14]
labels=["Dk","Orta","Yksek","Kritik"]
colors=['#13EAC9',"#069AF3","#FFFF00","#DDA0DD"]
plt.pie(area,labels=labels,colors=colors,shadow = 'true', autopct = '%1.1f%%', explode=[0,0,0.1,0.8])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,8))
data=[100,25,45,65]
labels=["Kimya","Fizik","Biyoloji","Matematik"]
colors=["#C1F80A","#C0CA33","#FF7F50","#15B01A"]
plt.pie(data,labels=labels,colors=colors ,shadow='true')
my_circle=plt.Circle((0,0),0.7,color="white")
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
plt.figure(figsize=(20,20))
data=[15,26,32,44]
marka_isimleri=["IBM","Microsoft","Apple","Dell"]
durum=["zld","ptal edildi","Askya alnd.","Atand"]
sayac=["Dk","Orta","Yksek","Kritik"]
colors=["#DDA0DD","#C0CA33","#13EAC9","#069AF3"]
my_circle=plt.Circle((0,0),0.7, color="white")

plt.subplot(1,3,1)
plt.pie(data,labels=marka_isimleri,colors=colors,shadow="true")
my_circle=plt.Circle((0,0),0.7, color="white")
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.legend()

plt.subplot(1,3,2)
plt.pie(data,labels=durum,colors=colors,shadow="true")
my_circle=plt.Circle((0,0),0.7, color="white")
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.legend()


plt.subplot(1,3,3)
plt.pie(data,labels=sayac,colors=colors,shadow="true")
my_circle=plt.Circle((0,0),0.7, color="white")
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.legend()

plt.show()
```
