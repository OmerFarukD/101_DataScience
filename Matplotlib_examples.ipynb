```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
```

```python
plt.figure(figsize=(10,6))
x=np.linspace(0,10,2000)
plt.plot(x,np.sin(x) ,label="Sin(x)")
plt.plot(x,np.cos(x),label="Cos(x)")
plt.xlabel("X" ,fontsize=18)
plt.ylabel("Y" ,fontsize=18)
plt.title("$Sin(x) $ $  &  $ $ Cos(x) $", fontsize = 14)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
x=np.array([1,2,3,4,5,6,7,8])
y1=np.power(x,2)
y2=np.power(x,4)
plt.plot(x,y1 ,"b-",label="$y1=x^2$")
plt.plot(x,y2 , color="red", linestyle="--",label="$y2=x^4$")
plt.xlabel("X ekseni")
plt.ylabel("y ekseni")
plt.legend(loc="upper left" ,fontsize="large")
plt.show()
```

```python
x=np.linspace(0,10,2000)
plt.figure(figsize=(10,6))
plt.plot(x,np.sin(x) ,label="Sin(x)")
plt.plot(x,np.cos(x),label="Cos(x)")
plt.fill_between(x,0,np.sin(x),color="yellow")
plt.fill_between(x,0,np.cos(x),color="red")
plt.xlabel(r"$X$",fontsize=18)
plt.ylabel(r"$Y$",fontsize=18)
plt.title("$Sin(x) $ $ & $ $ Cos(x) $", fontsize = 14)
plt.legend( loc="lower left")
```

```python
x=np.arange(-50,50)
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.plot(x,np.power(x,2), label="$y=x^2$")
plt.legend(loc="upper center")
plt.subplot(1,2,2)
plt.plot(x,np.tan(x),label="$y=tan(x)$")
plt.legend(loc="upper right")
plt.show()
```

```python
x=np.arange(-20,20)
y=np.power(x,2)
y1=np.power(x,3)
y2=np.log10(x)
y3=np.exp(x)
fig1,ax1=plt.subplots(nrows=2,ncols=2,figsize=(20,20))
ax1[0,0].plot(x,y, color="red")
ax1[0,0].set_title("$y=x^2$")
ax1[0,0].set_xlabel(r"$X$ label")
ax1[0,0].set_ylabel(r"$y$ label")

ax1[0,1].plot(x,y1, color="green")
ax1[0,1].set_title("$y=x^3$")
ax1[0,1].set_xlabel(r"$X$ label")
ax1[0,1].set_ylabel(r"$y$ label")

ax1[1,0].plot(x,y2, color="yellow")
ax1[1,0].set_title("$y=log(x)$")
ax1[1,0].set_xlabel(r"$X$ label")
ax1[1,0].set_ylabel(r"$y$ label")

ax1[1,1].plot(x,y3, color="blue")
ax1[1,1].set_title("$y=exp(x)$")
ax1[1,1].set_xlabel(r"$X$ label")
ax1[1,1].set_ylabel(r"$y$ label")
```
